```python
import pandas as pd
#convert parquet file into csv
parquet_file = '~/Downloads/yellow_tripdata_2021-01.parquet'
df = pd.read_parquet(parquet_file)
df.to_csv('~/Downloads/yellow_tripdata_2021-01.csv')
```

```python
df
```

```python
##DDL of the schema to Postgres
print(pd.io.sql.get_schema(df, name= 'yellow_taxi_data'))
```

```python
#create connection to postgres and tell postgres to generate the schema above
from sqlalchemy import create_engine
```

```python
engine = create_engine('postgresql://root:root@localhost:5431/ny_taxi')
```

```python
engine.connect()
```

```python
##DDL of the schema to Postgres
#once engine is connected, we can add the engine into it
#the table data type is written in sql language type
#NOTE this in parquet
print(pd.io.sql.get_schema(df, name= 'yellow_taxi_data', con=engine))
```

```python
#reading the parquet version with large dataset in chunks
csv_file = '~/Downloads/yellow_tripdata_2021-01.csv'
df_iter = pd.read_csv(csv_file, iterator=True, chunksize=100000)
```

```python
df = next(df_iter)
```

```python
df.tpep_pickup_datetime = pd.to_datetime(df.tpep_pickup_datetime)
df.tpep_dropoff_datetime = pd.to_datetime(df.tpep_dropoff_datetime)
```

```python
df.head(n=1)
```

```python
#this is the csv file version
#parquet file have better schema detection, csv need to manually change datetime
print(pd.io.sql.get_schema(df, name= 'yellow_taxi_data'))
```

```python
#if_exists='replace' is same as CREATE yellow_taxi_data TABLE in sql
df.to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
#if_exists='append' is same as INSERT query of data to db
%time df.to_sql(name='yellow_taxi_data', con=engine, if_exists='append')
```
